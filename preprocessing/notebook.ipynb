```python
import pandas as pd
import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import wordpunct_tokenize
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
df = pd.read_csv(
    "C:/Users/Mouli/TokenizationReview_Sentiment/data/reviews.csv",
    header=None,
    names=["review"]
)

print(df.head())
```

```python
stop_words = set(stopwords.words('english'))

def preprocess_steps(text):
    
    # 1. Lowercase
    lower_text = text.lower()
    
    # 2. Tokenization
    tokens = wordpunct_tokenize(lower_text)
    
    # 3. Remove punctuation
    no_punct_tokens = [t for t in tokens if t not in string.punctuation]
    
    # 4. Remove stopwords
    final_tokens = [t for t in no_punct_tokens if t not in stop_words]
    
    return lower_text, tokens, no_punct_tokens, final_tokens


# Apply
result = df["review"].apply(preprocess_steps).tolist()

# Create columns
df[["lowercase", "tokens", "no_punct", "final_tokens"]] = pd.DataFrame(result)

print(df)
```
